```python
import pandas as pd
```

```python
df = pd.read_excel('multum_alerts_smh_deid_2019_05_23.xlsx')
pd.to_datetime(df['ALERT_DATE_TIME'], format='%Y%m%d', errors='ignore')
df.fillna(0, inplace=True)
df.head()
```

```python
for idx, row in df.iterrows():
    sql = "INSERT INTO modelpatient.upmc_alerts "
    sql += "(facility,unit,room,bed,alert_date_time,drug,alert_type, "
    sql += "alert_severity, override_reason, physician_id) VALUES ("
    sql += "'" + row['FACILITY'] + "', "
    sql += "'" + row['UNIT'] + "', "
    sql += "'" + str(row['ROOM']) + "', "
    sql += "'" + str(row['BED']) + "', "
    sql += "'" + str(row['ALERT_DATE_TIME']) + "', "
    sql += "'" + row['DRUG'] + "', "
    sql += "'" + row['ALERT_TYPE'] + "', "
    sql += "'" + str(row['ALERT_SEVERITY']) + "', "
    sql += "'" + str(row['OVERRIDE_REASON']) + "', "
    sql += "'" + str(row['ORDERING_PHYSICIAN_DEID']) + "'); "
    
    print(sql)
```
