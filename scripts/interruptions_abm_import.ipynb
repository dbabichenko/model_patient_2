```python
import pandas as pd
import json
```

```python
df = pd.read_csv('Interruptions-experiment-table.csv')
df.head()
```

```python
cols = ['run_number', 
'number_of_doctors', 
'number_of_patients', 
'step', 
'count_doctors', 
'count_patients', 
'count_interrupters_alerts', 
'count_interrupters_devices', 
'count_interrupters_people', 
'count_interrupters_alarms', 
'count_doctors_interrupted', 
'count_doctors_consequence_resume_prev', 
'count_doctors_consequence_switch_task', 
'list_active_interrupters', 
'type_active_interrupters', 
'current_interrupt_doctors', 
'conseq_interrupt_doctors'
]
```

```python
df.columns = cols
df.head()
```

```python
df[df.columns[1:]]
```

```python
q = "number_of_doctors == 1 & number_of_patients == 2"
df_filtered = df.query(q)
```

```python
df_filtered.shape
```

```python
data = []
for idx, row in df_filtered.iterrows():
    data_row = {}
    for col in df_filtered.columns[3:]:
        data_row[col] = row[col]
        data.append(data_row)
#print(data)
```

```python
with open('interruptions.json', 'w') as fp:
    json.dump(data, fp)
```
